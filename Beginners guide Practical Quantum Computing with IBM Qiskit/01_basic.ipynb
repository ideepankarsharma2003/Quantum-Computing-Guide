**Qiskit** (pronounced kiss-kit) is comprised of four elements, each of which has a specific functionality and role that can be leveraged based on the areas that you wish to experiment with. 

These elements are:

- **Terra (Earth)** : composing quantum programs at the level of circuits and pulses with code foundation
- **Aer (Air)** : accelerating development via simulators and noise models
- **Ignis (Fire)** : addressing noise and errors
- **Aqua (Water)** : building algorithms and applications

```python
import qiskit 
qiskit.__version__
```

```python
from qiskit_aer import Aer
from qiskit import QuantumCircuit, transpile
from qiskit.visualization import plot_bloch_multivector
```

```python
# create a quantum circuit with single qubit, default initial state: |0> ---> [1, 0]
qc= QuantumCircuit(1)

qc.x(0) # apply X-gate on the |0> qubit

qc.draw("mpl")

# get the backend for the circuit
backend= Aer.get_backend("statevector_simulator")


# execute the circuit using backend
compiled_circuit = transpile(qc,backend)
out = backend.run(compiled_circuit).result().get_statevector()
plot_bloch_multivector(out)
print(f"state_vector: {out}")
```
